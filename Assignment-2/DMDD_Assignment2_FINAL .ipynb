```python
#Code for web scrapping IMDB_movies using BeutifulSoup

from bs4 import BeautifulSoup
import requests


res=requests.get("https://www.imdb.com/search/title?title_type=feature&release_date=2018-01-01,2019-02-28&user_rating=7.3,&count=100")

soup=BeautifulSoup(res.text,'html.parser')
print(soup)
    
id_check = soup.find(id ="main")
movie_container = id_check.find_all(class_ ="lister-item mode-advanced")
len(movie_container)

#list to store scraped value data in:
movie_name = []
year_release = []
imdb_ratings = []
metascores = []
votes = []
runtime = []
genre = []
director_name = []
star_cast = []



#extract data from individual movie container
for container in movie_container:
    
   
    #if movie has Metascore, then extract:
    if container.find("div", class_ = "ratings-metascore") is not None:
        
        
        #the movie_name
        name = container.h3.a.text
        movie_name.append(name)
        
        #the year od release
        release = container.find("span", class_ = "lister-item-year text-muted unbold").text
        year_release.append(release)
        
        #the ratings for the movies
        ratings = float(container.strong.text)
        imdb_ratings.append(ratings)
        
        #the metascores
        meta = container.find("span", class_ = "metascore").text
        metascores.append(int(meta))
        
        #the votes
        vote = container.find("span", attrs = {"name":"nv"})['data-value']
        votes.append(int(vote))

        #the runtime
        run = container.find("span", class_ ="runtime").text
        runtime.append(run)
        
        #the genre
        gen = container.find("span", class_ ="genre").text
        genre.append(gen)
        
        #fetching all <p> tags
        content = container.find_all("p")
        
        #subsetting all the <a> tags in 3rd <p> tag
        content_2 = content[2].find_all("a")
        
        #the director
        director = content_2[0].text
        director_name.append(director)
        
        
        #extracting artists names
        temp = []
        for i in range(len(content_2)-1):
            temp.append(content_2[i].text)
        star_cast.append(temp)
```

```python
import pandas as pd

movie_df = pd.DataFrame({"movie_names":movie_name,
                        "year_release":year_release,
                        "imdb_ratings":imdb_ratings,
                        "metscores":metascores,
                        "votes":votes,
                        "runtime":runtime,
                        "genre":genre,
                        "director_name": director_name,
                        "star_cast": star_cast,
                       })
```

```python
import re

def split_it(year):
    return re.findall('(\d\d\d\d)', year)
movie_df['year_release'] = movie_df['year_release'].apply(split_it)
movie_df['genre'] = movie_df['genre'].str.strip("\n")

for index,row in movie_df.iterrows():
    movie_df.at[index,'star_cast'] = ('|'.join(map(str,row["star_cast"])))

for index,row in movie_df.iterrows():
    movie_df.at[index,'year_release'] = (' '.join(map(str,row["year_release"])))
    
movie_df
```

```python
#Instagram Entity (Consumers)

from datetime import datetime
from itertools import dropwhile, takewhile

import instaloader
import pandas as pd

L = instaloader.Instaloader()

#creating a list to store all the attribute values
movie_names = []
Caption = []
Caption_Hashtags = []
Caption_Hashtags_Each =[]
Timestamp = []
User_ID = []
Likes = []

# just make a list of all the hash tags that you want
#for i in list: 
#create a loop and pass the hashtag to the instaloader function


HashtagsList=['Aquaman','AStarIsBorn', 'CaptainMarvel','IsntItRomantic','Alita','Creed2','Incredibles2','Deadpool2','GullyBoy','DetectivePikachu']

all_posts = []


#Loop to get all the posts by passing the Names of movies in a list
for i in HashtagsList:
    counter = 0
    posts = L.get_hashtag_posts(i)
    SINCE = datetime(2019, 3, 5)
    for post in takewhile(lambda p: p.date > SINCE, posts):
        if counter == 4:
            break
        movie_names.append(i)
        Caption.append(post.caption)
        Caption_Hashtags.append(post.caption_hashtags)
        Timestamp.append(post.date)
        User_ID.append(post.owner_username)
        Likes.append(post.likes)
        counter+=1
```

```python
movies_insta_df = pd.DataFrame({"movie_names":movie_names,
                        "Caption":Caption,
                        "Caption_Hashtags":Caption_Hashtags,
                        "Timestamp":Timestamp,
                        "User_ID":User_ID,
                        "Likes":Likes })
```

```python
import re

def split_it1(name_movie):
    return re.sub(r'([a-z](?=[A-Z])|[A-Z](?=[A-Z][a-z]))', r'\1 ', name_movie)
movies_insta_df['movie_names'] = movies_insta_df['movie_names'].apply(split_it1)

#To separate the hashtags with a |
for index,row in movies_insta_df.iterrows():
    movies_insta_df.at[index,'Caption_Hashtags'] = ('|'.join(map(str,row["Caption_Hashtags"])))
```

```python
import sqlite3
conn = sqlite3.connect("DMDD_Assignment2_IP")
c = conn.cursor()
```

```python
#Creating table for Movies Entity
c.execute("create table insta_movies_tab(movies_name text, Caption text, Caption_Hashtags text, Timestamp timestamp, User_ID text, likes integer )")

#Creating table for Social Media:Instagram(Consumers) Entity
c.execute("create table imdb_movies(movie_names text, year_release text, imdb_ratings integer, metascores integer, votes integer, runtime varchar, genre text, director_name text, star_cast text )")

#Creating table for Social Media:Twitter(Consumers) Entity
c.execute("create table twitter_movies(productionhouse_name text, User_id integer, Followers integer, Post text, Hashtags text, Posted_time timestamp, movie_names text, Post_ID integer )")

#Creating table for Twitter_movies Entity(Company)
c.execute("create table twitter_movies_post(User_Name text, User_id integer, User_Post text, Hashtags text, Timestamp timestamp, movie_names text )")

conn.commit()
```

```python
#Importing all the four dataframes to SQL

movie_df.to_sql('imdb_movies',
           con=conn,
           if_exists='replace')

movies_insta_df.to_sql('insta_movies_tab',
           con=conn,
           if_exists='replace')

twitter_movies.to_sql('twitter_movies',
           con=conn,
           if_exists='replace')


twitter_movies_post.to_sql('twitter_movies_post',
           con=conn,
           if_exists='replace')
```
