```python
#Importing all the Required Libraries

import json
from requests import get
import numpy as np
import pandas as pd
import requests
import re
```

```python
#First Source:Web Scraping
#Scraping data from the Website:'https://pokemondb.net/pokedex/all'

url = "https://pokemondb.net/pokedex/all"

response = get(url)
response

## importing Beautiful Soup
from bs4 import BeautifulSoup as bs

## python's in built library HTML parser
html_soup = bs(response.text,'html.parser')
type(html_soup)

pokemon_check = html_soup.find(id ="pokedex")
poke_container = pokemon_check.find_all('tr')
len(poke_container)

#Creating lists for each of the columns required to be placed in a DataFrame.

ls = []
TypesPresent = ["Normal", "Ice", "Fire", "Water", "Flying", "Electric" , "Grass", "Fighting", "Poison", "Ground", "Psychic",
               "Bug", "Dark", "Fairy", "Ghost", "Rock", "Dragon", "Steel"]
Name = []
Type1 = []
Type2 = []
Total = []
HP = []
Attack = []
Defence = []
SpAttack = []
SpDefence = []
Speed = []


for i in range(1, len(poke_container)):

#Sample Data of each Row    
#     ['"', '001 BulbasaurGrassPoison', '318', '45', '49', '49', '65', '65', '45', '"']

    ls = json.dumps(poke_container[i].text).split("\\n")
    
    Total.append(ls[2])
    HP.append(ls[3])
    Attack.append(ls[4])
    Defence.append(ls[5])
    SpAttack.append(ls[6])
    SpDefence.append(ls[7])
    Speed.append(ls[8])
    
    types = re.findall('[A-Z][^A-Z]*', ls[1])
    
    Name.append(types[0])
    
    if types[-1] in TypesPresent:
        Type1.append(types[-1])
    else :
        Type1.append(" ")
    if types[-2] in TypesPresent:
        Type2.append(types[-2])
    else :
        Type2.append("Not Available")
        
        
poke_df = pd.DataFrame({"Name":Name,
                        "Type1":Type1,
                        "Type2":Type2,
                        "Total":Total,
                        "HP":HP,
                        "Attack":Attack,
                        "Defence":Defence,
                        "Sp.Attack":SpAttack,
                        "Sp.Defence":SpDefence,
                        "Speed": Speed
                       })

pd.set_option('display.max_colwidth', -1)

poke_df
```

```python
#Converting the Web Sorce dataframe into a .CSV file

poke_df.to_csv(r'C:\Users\purva\Pokemon_webScrape.csv')
```

```python
#Cleaning the Data and Auditing it from the WebSource Dataframe

poke_df.isnull().sum()
```

```python
#Second Source of Data:Web API
#Fetching Data from the following APIs

baseUrl = "https://pogoapi.net/api/v1/"

stat = "pokemon_stats"
home = "shiny_pokemon"
types = "pokemon_types"
maxCp = "pokemon_max_cp"
chargedMoves = "charged_moves"
fastMoves = "fast_moves"
pokemon_encounter = "pokemon_encounter_data"

extension = ".json"

stats_dict = requests.get(baseUrl+stat+extension).json()
types_dict = requests.get(baseUrl+types+extension).json()
maxCp_dict = requests.get(baseUrl+maxCp+extension).json()
encounter_data_dict = requests.get(baseUrl+pokemon_encounter+extension).json()

#Combining all dictionaries into one dictionary to conver it to a final list

pokemon_aggregate = zip(stats_dict,types_dict, maxCp_dict, encounter_data_dict)
pokemon_aggregate_list = list(pokemon_aggregate)

final_list = list()

for pokemon in pokemon_aggregate_list:
    temp = dict()
    for d,poke  in enumerate(pokemon):
        if d == 0:
            temp['base_attack'] = poke['base_attack']
            temp['base_defense'] = poke['base_defense']
            temp['base_stamina'] = poke['base_stamina']
            temp['pokemon_name'] = poke['pokemon_name']
        elif d == 1:
            temp['type'] = poke['type']
        elif d == 2:
            temp['max_cp'] = poke['max_cp']
        else:
            temp['attack_probability'] = poke['attack_probability']
            temp['base_capture_rate'] = poke['base_capture_rate']
            temp['base_flee_rate'] = poke['base_flee_rate']
            temp['max_pokemon_action_frequency'] = poke['max_pokemon_action_frequency']
            temp['min_pokemon_action_frequency'] = poke['min_pokemon_action_frequency']
            
    final_list.append(temp)


##Creating lists for each of the columns required to be placed in a DataFrame

base_attack = list()
pokemon_name = list()
base_defense = list()
base_stamina = list()
type1 = list()
type2 = list()
max_cp = list()
attack_probability = list()
base_capture_rate = list()
base_flee_rate = list()
max_pokemon_action_frequency = list()
min_pokemon_action_frequency = list()

for pokemon in final_list:
    base_attack.append(pokemon['base_attack'])
    base_defense.append(pokemon['base_defense'])
    base_stamina.append(pokemon['base_stamina'])
    for i,t in enumerate(pokemon['type']):
        if len(pokemon['type']) < 2:
            type2.append("Not Available")
        if i == 0:
            type1.append(t)
        elif i == 1:
            type2.append(t)
    max_cp.append(pokemon['max_cp'])
    pokemon_name.append(pokemon['pokemon_name'])
    attack_probability.append(pokemon['attack_probability'])
    base_capture_rate.append(pokemon['base_capture_rate'])
    base_flee_rate.append(pokemon['base_flee_rate'])
    max_pokemon_action_frequency.append(pokemon['max_pokemon_action_frequency'])
    min_pokemon_action_frequency.append(pokemon['min_pokemon_action_frequency'])
    
poke_api_df = pd.DataFrame({"Name":pokemon_name,
                        "Base_Attack":base_attack,
                        "Base_Defense":base_defense,
                        "Base_Stamina":base_stamina,
                        "Type1":type1,
                        "Type2":type2,
                        "Max_CP":max_cp,
                        "Attack_Probability":attack_probability,
                        "Base_Capture_Rate":base_capture_rate,
                        "Base_Flee_Rate": base_flee_rate,
                        "Max_Pokemon_Action_Frequency": max_pokemon_action_frequency,
                        "Min_Pokemon_Action_Frequency":min_pokemon_action_frequency
                       })

poke_api_df
```

```python
#Converting the Web API dataframe into a .CSV file

poke_api_df.to_csv(r'C:\Users\purva\Pokemon_webAPI.csv')
```
